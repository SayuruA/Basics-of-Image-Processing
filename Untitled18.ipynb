```python
def sobel(im):
    kernel = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])  # Sobel kernel for x-direction
    im_pad = np.zeros((im.shape[0]+2, im.shape[1]+2), dtype = np.int16)  # Zero-padding the image
    im_pad[1:-1, 1:-1] = im.astype(np.int16)

    # Create an output array to store the result
    result = np.zeros_like(im, dtype = np.int16)

    # Apply the kernel for one pixel at a time
    for i in range(1, im.shape[0]+1):
        for j in range(1, im.shape[1]+1):
            result[i-1, j-1] = np.sum(kernel * im_pad[i-1:i+2, j-1:j+2])

    # Clip values to stay within [0, 255] and convert back to uint8
    result = np.clip(result, 0, 255)
    print(result.min(),result.max())
    return result.astype(np.uint8)
```

```python
def sobel_improved(im):
    # Define the 1D Sobel kernels (separable)
    sobel_kernel_v = np.array([1, 2, 1])  # Vertical 1D kernel (for smoothing)
    sobel_kernel_h = np.array([1, 0, -1])  # Horizontal 1D kernel (for edge detection)

    # Initialize empty arrays to store intermediate and final results
    sobel_x_intermediate = np.zeros_like(im, dtype=np.float64)
    sobel_x = np.zeros_like(im, dtype=np.float64)

    # Padding the image to handle the edges (for 1D vertical filter)
    padded_img = np.pad(im, ((1, 1), (0, 0)), mode='constant')

    # Step 1: Apply vertical 1D kernel (along the rows)
    for i in range(1, im.shape[0]+1):
        for j in range(0, im.shape[1]):
            # Apply the vertical kernel to the 3-pixel neighborhood in the column
            sobel_x_intermediate[i-1, j] = np.sum(padded_img[i-1:i+2, j] * sobel_kernel_v)

    # Padding the intermediate result for horizontal filtering
    padded_intermediate = np.pad(sobel_x_intermediate, ((0, 0), (1, 1)), mode='constant')

    # Step 2: Apply horizontal 1D kernel (along the columns)
    for i in range(0, im.shape[0]):
        for j in range(1, im.shape[1]+1):
            # Apply the horizontal kernel to the 3-pixel neighborhood in the row
            sobel_x[i, j-1] = np.sum(padded_intermediate[i, j-1:j+2] * sobel_kernel_h)

    # Convert the result back to uint8 format, normalizing the values to 0-255
    sobel_x = np.uint8(np.clip(np.abs(sobel_x), 0, 255))

    return sobel_x
```

```python
sobel_im = sobel_improved(im_sobel)
# show sobel_im
plt.imshow(sobel_im, cmap = "gray")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to compute normalized SSD
def normalized_ssd(original, zoomed):
    diff = original.astype(np.float32) - zoomed.astype(np.float32)
    squared_diff = np.square(diff)
    return np.mean(squared_diff)

# Function to zoom an image using nearest-neighbor and bilinear interpolation
def zoom_image(image, scale_factor, method='nearest'):
    if method == 'nearest':
        # Use OpenCV's resize with INTER_NEAREST for nearest-neighbor interpolation
        zoomed_image = cv2.resize(image, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_NEAREST)
    elif method == 'bilinear':
        # Use OpenCV's resize with INTER_LINEAR for bilinear interpolation
        zoomed_image = cv2.resize(image, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_LINEAR)
    return zoomed_image

# Function to test zooming and compute SSD
def test_zoom(image, zoom_out_factor, zoom_in_factor=4):
    # Step 1: Zoom out the image
    downscaled_image = cv2.resize(image, (0, 0), fx=1/zoom_out_factor, fy=1/zoom_out_factor, interpolation=cv2.INTER_LINEAR)

    # Step 2: Zoom back in using both methods
    nn_zoomed = zoom_image(downscaled_image, zoom_in_factor, method='nearest')
    bilinear_zoomed = zoom_image(downscaled_image, zoom_in_factor, method='bilinear')

    # Step 3: Compute SSD between original and zoomed images
    nn_ssd = normalized_ssd(image, nn_zoomed)
    bilinear_ssd = normalized_ssd(image, bilinear_zoomed)

    return nn_zoomed, bilinear_zoomed, nn_ssd, bilinear_ssd

# Load your image
image_path = "/content/drive/MyDrive/a1images/a1q5images/im01small.png"  # Update with your image path
image = cv2.imread(image_path)

# Convert image from BGR (OpenCV default) to RGB for visualization in matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Test the zooming methods with a downscale factor of 4
nn_zoomed_image, bilinear_zoomed_image, nn_ssd, bilinear_ssd = test_zoom(image_rgb, zoom_out_factor=4)

# Print the SSD results
print(f"Nearest-Neighbor SSD: {nn_ssd}")
print(f"Bilinear Interpolation SSD: {bilinear_ssd}")

# Display the original and zoomed images
plt.figure(figsize=(10, 10))
plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(nn_zoomed_image)
plt.title('Nearest-Neighbor Zoom')

plt.subplot(1, 3, 3)
plt.imshow(bilinear_zoomed_image)
plt.title('Bilinear Zoom')

plt.show()

```
